```python
from pyspark.sql.types import *
```

```python
def py_replaceNull(origi):
    if origi == None:
        result = None
    else:
        result = origi.replace('%','')
    return result
```

```python
from pyspark.sql import functions as sparkf
```

```python
spark_replceNull = sparkf.udf(lambda x: py_replaceNull(x))
```

```python
noPercent_df = selected_df.withColumn('int_rate',\
                       spark_replceNull(sparkf.col('int_rate')))\
.withColumn('int_rate',sparkf.col('int_rate').cast(DoubleType()))
```

```python
noPercent_df.printSchema()
```

```python
#ระบุว่า Nulls อยู่ที่ row ไหน
from pyspark.sql import functions as sparkf
from functools import reduce

nullAnalysis_df = noPercent_df.filter(reduce(lambda a1, a2: a1 | a2,\
                     (sparkf.col(c).isNull() \
                      for c in noPercent_df.columns)))
```

```python
nullAnalysis_df.printSchema()
```

```python
nullAnalysis_df.describe().toPandas().transpose()
```

```python
nullAnalysis_df.count()
```

```python
nullRow_num = nullAnalysis_df.count()
```

```python
nullAnalysis_df.groupBy('int_rate','bc_open_to_buy').count()\
.orderBy('count',ascending = False)\
.withColumn('percent',sparkf.col('count')/nullRow_num).show(1000)
```

```python
double_df = noPercent_df.withColumn('bc_open_to_buy',\
                       sparkf.col('bc_open_to_buy').cast(DoubleType()))
```

```python
avgBC_OPEN_TO_BUY = double_df.na.drop().agg(sparkf.avg(sparkf.col('bc_open_to_buy')))\
.collect()[0][0]
```

```python
no_null_df = double_df.withColumn('bc_open_to_buy'\
    ,sparkf.when(sparkf.col('bc_open_to_buy').isNull()\
    ,avgBC_OPEN_TO_BUY).otherwise(sparkf.col('bc_open_to_buy')))
```
